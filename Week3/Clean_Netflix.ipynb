```python
df['release_year']=pd.to_numeric(df['release_year'])
df['seasons']=pd.to_numeric(df['seasons'])
#Removed all those which didn't have a release year
df = df.dropna(subset=['release_year', 'type'])
df = df[df['release_year'] > 1900]
```

#Now Adding Features

```python
#Content age
current_year=pd.Timestamp.now().year
df['content_age']=current_year-df['release_year']
#Release Decade
df['release_decade'] = (df['release_year'] // 10) * 10
#Movie/TV ratio
yearly_type_counts = df.pivot_table(
    index='release_year',
    columns='type',
    aggfunc='size',
    fill_value=0
)

yearly_type_counts['movie_tv_ratio'] = (
    yearly_type_counts['MOVIE'] / yearly_type_counts['SHOW']
)
yearly_type_counts
```
