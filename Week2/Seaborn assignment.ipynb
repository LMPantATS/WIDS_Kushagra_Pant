```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='darkgrid')
```

##  Load Datasets

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')

tips.head()
```

##  Relational Plots

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker')
plt.title('Total Bill vs Tip')
plt.show()
```

##  Categorical Plots

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title('Average Tip by Day')
plt.show()
```

##  Distribution Plots

```python
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

##  Regression Plot

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.title('Regression: Total Bill vs Tip')
plt.show()
```

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
sns.set_theme(style='darkgrid')
plt.figure(figsize=(8, 5))
#change grid and color
sns.set_theme(style="whitegrid")
sns.set_palette("Set3")
sns.lineplot(data=iris, x='species', y='sepal_length', label='Sepal Length')#, hue='', style='smoker')
sns.lineplot(data=iris, x='species', y='sepal_width', color='r', label='Sepal Width')
sns.lineplot(data=iris, x='species', y='petal_length', color='b', label='Petal Length')
sns.lineplot(data=iris, x='species', y='petal_width', color='orange', label='Petal Width')
plt.ylabel("Length")
plt.xlabel("Species")
plt.title("Average Iris Measurements by Species")

#sns.histplot(data=iris, x="species", y="sepal_length", kde=True)

plt.show()
```

**This plot shows average sepal length, sepal width, petal length and petal width across various species. Petal length and petal width vary more significantly between species compared to sepal measurements.It means that petal dimensions are main indicator of species. Sepal length and sepal width show relatively smaller differences across species, making them a weak indicator for species.**
